## Financially Healthy Companies:

- **Understand key business relationships**: Metrics like `total_assets` should generally be greater than `current_assets`. Similarly, `gross_profit` should be derived from `net_sales` and `cost_of_goods_sold`. These relationships should guide financial analysis.


### Key Characteristics of Healthy Companies:
- **Gross Profit > 0**: Derived from `net_sales - cost_of_goods_sold`.
- **Positive EBIT, EBITDA, and Net Income**: Reflect profitability and sound financial performance.
- **Total Liabilities ≤ Total Assets**: Indicates positive equity and financial stability.
- **Reasonable Market Value**: Linked to profitability metrics such as `net_income` or `ebitda`.

### Business Logic:
1. **Logical Relationships**:
   - `gross_profit = net_sales - cost_of_goods_sold`
   - `ebit = gross_profit - total_operating_expenses - depreciation_and_amortization`
   - `ebitda = ebit + depreciation_and_amortization`
   - `net_income = ebit - total_long_term_debt`
   - `total_liabilities` should not exceed `total_assets`.
   
2. **Realistic Ranges**: Values are set within business-appropriate ranges while maintaining logical relationships.
   
3. **Introduce Variability**: Financial metrics exhibit realistic variability within reasonable limits, ensuring that performance is dynamic yet consistent with the company's healthy standing.

### Outcome:
This approach ensures that financial metrics for **healthy companies** maintain logical relationships. Metrics like `net_sales`, `gross_profit`, `ebit`, and `total_assets` are consistent with sound business practices. Variations are introduced to reflect normal fluctuations while preserving the company's financial health.

## Financial Metrics for Companies

This function calculates financial metrics for companies, accounting for both healthy and distressed companies. The financial behavior of distressed companies is modeled through lower profitability, higher liabilities, and reduced market value, while healthy companies maintain more favorable financial conditions.

### Key Adjustments for Distressed Companies:
- **Revenue Growth**: Distressed companies experience negative or slower revenue growth compared to healthy companies. Growth rates are adjusted based on company size and financial health.
- **Gross Margins**: Distressed companies have lower or even negative gross margins, reflecting poor profitability.
- **Operating Expenses**: Higher operating expenses are applied to distressed companies, leading to decreased profitability.
- **Net Income**: For distressed companies, net income is low or negative, reflecting financial strain.
- **Assets and Liabilities**: Distressed companies have higher liabilities relative to their assets, with a cap ensuring liabilities do not exceed assets.
- **Retained Earnings**: Negative retained earnings are calculated for distressed companies, reflecting accumulated losses.
- **Market Value**: The market value for distressed companies declines, reflecting their financial instability.

### Business Logic:
1. **Revenue**: Generated using growth rates that vary based on company size and distress status.
2. **Profitability Metrics**: `gross_profit`, `ebit`, `ebitda`, and `net_income` are calculated using industry-specific expense ratios and accounting for distress factors.
3. **Assets and Liabilities**: Total liabilities are capped to ensure they remain below total assets, while distressed companies face higher liability ratios.
4. **Realistic Variability**: Financial metrics are adjusted with variability to reflect the dynamic nature of company performance.

### Outcome:
This method ensures that the financial metrics for both healthy and distressed companies are consistent with business logic. Distressed companies exhibit key financial challenges, such as negative profitability, high liabilities, and lower market value, while healthy companies maintain stable financial performance.


```python
# Modify the function to generate financial metrics for distressed companies
def generate_financials(company, distressed=False):
    num_periods = len(company)
    industry = company['industry'].iloc[0]
    size_category = company['size_category'].iloc[0]
    
    # Set base revenue ranges based on company size
    if size_category == 'Small':
        revenue_lower, revenue_upper = 50000, 500000
    elif size_category == 'Medium':
        revenue_lower, revenue_upper = 500000, 1500000
    else:  # Large
        revenue_lower, revenue_upper = 1500000, 2000000

    # Generate initial revenues within the specified range
    initial_revenue = np.random.normal(revenue_lower, revenue_upper)

    def get_growth_rate(size_category, distressed=False):
        if distressed:
            # Distressed companies may have declining revenues
            return np.random.normal(-0.1, 0.15, num_periods).clip(-0.3, 0.1)
        if size_category == 'Small':
            mean, std = 0.15, 0.1  # Higher growth potential for non-distressed
        elif size_category == 'Medium':
            mean, std = 0.08, 0.05
        else:
            mean, std = 0.03, 0.01 # Lower growth for large companies
            
        return np.random.normal(mean, std, num_periods).clip(-0.2, 0.3)

    # Apply random growth rates (negative for distressed companies)
    growth_rates = get_growth_rate(size_category, distressed=distressed)
    revenues = initial_revenue * np.cumprod(1 + growth_rates)
    revenues = np.clip(revenues, revenue_lower, revenue_upper)

    # Gross margins for distressed companies are lower or negative
    gross_margin_lower, gross_margin_upper = industries[industry]['gross_margin']
    if distressed:
        gross_margins = np.random.normal(-0.1, 0.1, num_periods).clip(-0.5, 0.2)  # Negative or low margins
    else:
        gross_margins = np.random.normal(gross_margin_lower, gross_margin_upper, num_periods)

    gross_profits = revenues * gross_margins

    # Higher operating expenses for distressed companies
    industry_expense_factors = {
        'Technology': 0.2,
        'Healthcare': 0.18,
        'Finance': 0.1,
        'Retail': 0.12,
        'Manufacturing': 0.15
    }
    if distressed:
        operating_expenses_pct = np.random.normal(0.4, 0.1, num_periods)  # High operating expenses
    else:
        operating_expenses_pct = np.random.normal(industry_expense_factors[industry], 0.05, num_periods)

    total_operating_expenses = revenues * operating_expenses_pct

    # Depreciation and amortization
    depreciation_and_amortization = np.random.lognormal(mean=7, sigma=1, size=num_periods)

    # EBIT and EBITDA - may be negative for distressed companies
    ebit = gross_profits - total_operating_expenses - depreciation_and_amortization
    ebitda = ebit + depreciation_and_amortization

    # For distressed companies, net income can be very low or negative
    if distressed:
        net_income_margins = np.random.normal(-0.05, 0.05, num_periods).clip(-0.3, 0.1)
    else:
        net_income_margins = np.random.normal(0.05, 0.15, num_periods)

    net_income = revenues * net_income_margins

    # Total assets and liabilities - high debt and low assets for distressed companies
    total_assets = revenues * np.random.normal(0.8, 1.5, num_periods)
    if distressed:
        total_liabilities = total_assets * np.random.normal(0.7, 1.0, num_periods)  # High debt for distressed companies
    else:
        total_liabilities = total_assets * np.random.normal(0.3, 0.6, num_periods)

    # Ensure assets > liabilities
    total_liabilities = np.minimum(total_liabilities, total_assets * 0.9)

    # Current assets and current liabilities
    current_assets = total_assets * np.random.normal(0.4, 0.6, num_periods)
    total_current_liabilities = total_liabilities * np.random.normal(0.4, 0.6, num_periods)

    # Inventory and receivables
    inventory = np.random.normal(5000, 50000, num_periods)
    total_receivables = revenues * np.random.normal(0.1, 0.3, num_periods)

    # Retained earnings - negative for distressed companies
    if distressed:
        retained_earnings = np.cumsum(net_income) * np.random.normal(-0.2, 0.5)
    else:
        retained_earnings = np.cumsum(net_income) * np.random.normal(0.5, 1.0)

    # Market value - declines for distressed companies
    if distressed:
        market_value = net_income * np.random.normal(1, 3, num_periods).clip(0, None)
    else:
        market_value = net_income * np.random.normal(5, 15, num_periods)

    # Assign generated data to the company DataFrame
    company['total_revenue'] = revenues
    company['net_sales'] = revenues
    company['gross_profit'] = gross_profits
    company['total_operating_expenses'] = total_operating_expenses
    company['depreciation_and_amortization'] = depreciation_and_amortization
    company['ebit'] = ebit
    company['ebitda'] = ebitda
    company['net_income'] = net_income
    company['total_assets'] = total_assets
    company['total_liabilities'] = total_liabilities
    company['current_assets'] = current_assets
    company['total_current_liabilities'] = total_current_liabilities
    company['inventory'] = inventory
    company['total_receivables'] = total_receivables
    company['retained_earnings'] = retained_earnings
    company['market_value'] = market_value
    company['cost_of_goods_sold'] = revenues - gross_profits
    company['total_long_term_debt'] = total_liabilities - total_current_liabilities

    # Ensure no negative values where inappropriate
    numeric_cols = company.select_dtypes(include=[np.number]).columns
    company[numeric_cols] = company[numeric_cols].clip(lower=0)

    return company

# Apply the function to each company with some companies being distressed
distressed_company_ids = np.random.choice(company_attributes['company_id'], size=int(0.15 * num_companies), replace=False)
df = df.groupby('company_id', group_keys=False).apply(lambda x: generate_financials(x, distressed=x['company_id'].iloc[0] in distressed_company_ids))

# Display the first few rows
df.head()
```
